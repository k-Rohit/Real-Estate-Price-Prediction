```python
# dropping unecessary columns
flats.drop(['link','property_id'],axis=1,inplace=True)
```

```python
# renaming the area column
flats.rename(columns={'area':'price_per_sqft'},inplace=True)
```

```python
# society column
flats.society.value_counts()
```

```python
# No of unqiue society values
flats['society'].nunique()
```

```python
# removing rows with in which there is rating specified in the value
import re
flats['society'] = flats['society'].apply(lambda name: re.sub(r'\d+(\.\d+)?\s?★', '', str(name)).strip()).str.lower()
```

```python
flats['price'].value_counts() # There is a value price on reuqest so got to remove it
```

```python
flats = flats[flats['price'] != 'Price on Request']
```

```python
def clean_price_col(price):
  if type(price) == float:
    return price
  temp_list = price.split(" ")
  price_num = float(temp_list[0])
  unit = temp_list[1].lower()
  if unit == 'lac':
    price_num = price_num/100
  elif unit == 'crore':
    price_num = price_num
  return price_num
```

```python
flats['price'] = flats['price'].apply(clean_price_col)
```
